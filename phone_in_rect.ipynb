```python
if not os.path.exists('./dataset'):
    os.mkdir("./dataset")  # создание папки в текущем каталоге
```

```python
split_list = []

for file_ in os.listdir('./dataset_phone'):
    split_list.append(file_.split("."))

for filename in split_list:
    if filename[1] in ["jpg", "jpeg", "png"]:
        name = ".".join(filename)
        img = cv2.imread("./dataset_phone/{}".format(name), cv2.IMREAD_UNCHANGED)
        json_file = "./dataset_phone/{}.json".format(filename[0])
        with open(json_file, 'r', encoding='utf-8') as f:
            coord = json.load(f)
        res_img = cv2.rectangle(img, (coord[1][0], coord[1][1]), (coord[1][2], coord[1][3]), (0, 255, 255), 5)
        name_file = './dataset/{}.jpg'.format(random.randint(0, 1000))
        cv2.imwrite(name_file, res_img) # запись изображений в папку
```

```python
img_list = []

for file_ in os.listdir('./dataset'):
  img_list.append(file_)

num = np.ceil(np.sqrt(len(img_list))).astype('int')

fig, ax = plt.subplots(num, num, figsize=(20,20))

for i in range(len(img_list)): 
    img = cv2.imread("./dataset/{}".format(img_list[i]), cv2.IMREAD_UNCHANGED)
    ax[i//num][i%num].imshow(img)
    ax[i//num][i%num].axis('off')

plt.show()
```
